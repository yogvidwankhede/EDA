```python
new = ds["Age"].str.split(" ", n = 3, expand = True)

ds["First"]= new[0]
 
ds["Second"]= new[1]

ds["Third"]= new[2]

ds["Fourth"]= new[3]

ds
```

```python
ds["First"] = pd.to_numeric(ds["First"])
```

```python
ds['Third'] = np.where(ds['Third'] == '1/2', '0.5', np.where(ds['Third'] == 'Nagpur', ds['Third'] == np.nan, ds['Third']))
```

```python
ds["Third"] = pd.to_numeric(ds["Third"])
```

```python
ds['Age'] = np.where(ds['Second']=='months', ds['First']/12, np.where(ds['Second']=='days', ds['First']/365, ds['First']))
```

```python
ds = ds.drop(columns=["First", "Second", "Third", "Fourth"])
```

```python
ds["Age"] = pd.to_numeric(ds["Age"])
```

```python
ds['Age group'] = np.where(ds['Age']>18, 'Adult', np.where(ds['Age']>12, 'Teenager', np.where(ds['Age']<1, 'Infant', 'Child')))
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = ds.select_dtypes(include=numerics)

len(newdf.columns)
```

```python
new = ds["Visit date"].str.split(" ", n = 1, expand = True)

ds["Date"]= new[0]
 
ds["Time"]= new[1]

ds
```

```python
new = ds["Date"].str.split("/|-" , n = 2, expand = True)

ds["Day"]= new[0]
 
ds["Month"]= new[1]

ds["Year"]= new[2]

ds
```
